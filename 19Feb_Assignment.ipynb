```python
## Q1) Flask is web application framework in Python. It is very lightweight and easy to use framework used for developing 
#      Web-APIs.
```

```python
## Q2)
from flask import Flask

app = Flask(__name__)

@app.route("/hello")
def hello_world():
    return "Hello, World!!"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
## Q3) App routing is basically mapping a specific URl to the associated function which is intended to do some task.
#      We use app routes with different names in order to differentiate among functions when we call them.
```

```python
## Q4)
from flask import Flask


app = Flask(__name__)

@app.route("/Welcome")
def test1():
    return "Welcome to ABC Corporation"

@app.route("/")
def test2():
    return "Company Name: ABC Corporation <br>Location: India <br>Contact Detail: 999-999-9999"


if __name__=="__main__":
    app.run(host="0.0.0.0")

```

```python
## Q5) To build dynamic URL for a function in Flask we use url_for().
 
from flask import Flask,redirect,url_for


app = Flask(__name__)

@app.route("/Underage/<int:age>")
def test1():
    return "UNDERAGE!! You are not eligible to vote"

@app.route("/Overage/<int:age>")
def test2():
    return "You are eligible to vote"

##to check age
@app.route("/age/<int:age>")
def results(age):
    result=""
    if age<18:
        result="Underage"
    else:
        result="Overage"

    return(redirect(url_for(result,age=age)))
    

if __name__=="__main__":
    app.run(host="0.0.0.0")
```
